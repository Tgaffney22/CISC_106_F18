### Glossary

### You will not be tested on these definitions, but you will need to know them to understand this class.

**argument**: (fix) A value passed to a function.

**branch**: One of the possible paths during the **flow of execution**. 

**bug**: An error in a program that either throws an error or results in code that does unexpected things.

**body**: See **code block**

**boolean expression**: An **expression** that **evaluates** to either ```True``` or ```False```. These are used by ```if``` statements and ```while``` loops. 

**chained conditional** A conditional **branch** with more than two possible flows of execution. In Python these are written with ```if``` , ```elif``` , and ```else```. Here, the ```if``` and each ```elif``` is a different 'link' in the chain. Within a **chained conditional**, only a single **code block** will be run. Once a conditional **evaluates** to True, its block is run and all the other ``elif`` conditionals are skipped.

**code block**: (fix) Code that is grouped together. For example, the block of code that is run when an ```if``` statement **evaluates** to ```True```. In Python, this code will also be indented the same amount. 

**comment**: Written explanations of how the code works. A single-line comment starts with a '#' while multi-line comments are wrapped in tripe-quotations. (""" Pretend that is comment is spread over multiple lines. """

**comparison operator**: An **operator** that compares two values and **evaluates** to ```True``` or ```False```. They are: ==, !=, >, <, >=, and <=.

**conditional statement** (fix) -- Edit infinite loop?

**data type**: The type of data a **variable** holds. For instance: integers, floats, strings, boolean, and lists. The data type determinites what you can do with the data. 

**debugging**: Finding and fixing program errors.

**Docstring**: A formatted **comment** at the beginning of a function that explains what the function does.

**edge case**: When a slight change in value causes a big change in expected outcome. For example, if a 18-year-old has to pay a higher price for admission than a child who is 17 or younger. When reading requirements for a program, keep in mind how different values will change the program's **flow of execution**. 

**evaluate**: Run a the code step-by-step. This can include multiple steps in a single line. For instance, within a single line of code, the inner-most parenthesis are evaluated first. Example: ```3 * (2 + 8)``` becomes ```3 * 10``` after the ```(2 + 8)``` is evaluated. In the next step, ```3 * 10``` is evaluated to ```30```

**exception**: An error in your code

**expression**: A line of code consisting of **operators** and values that **evaluate** to a single value. For example, when the expression: ```my_int = (2 * 5) + 3``` is finished evaluating, the variable ```my_int``` contines the value ```13```

**flow of execution**: The order that lines in a program are run. **Statements** such as ```if``` and ```while``` change the **flow of execution** based on how many times they run. For example, if you have ```if```-```else``` statements that test to see if a user is old enough to vote, the **flow of execution** will take a different **branch** if the user is 17 versus if they are 18. 

**function**: (fix)

**generalize**: To replace code that does something specific with code that handles more cases. For example: When we rewrote our ```is_even()``` function that took a single integer and returned a boolean ```True``` if the integer was even with the function ```is_evenly_divisible()``` that took two argument and returned a boolean ```True``` if the first number was evenly divisible by the second number. 

**infinite loop**: A badly-formed ```while``` loop where the conditional statement never **evaluates** to ```False```. 

**initialization**: Create a variable with an intitial value. Example: ```an_int = 5``` initialzes the variable ```an_int``` and assigns it the value ```5```. Or ```user_input = input('Enter your age')``` intializes the variable ```user_input``` with whatever the user types into their keyboard.

**iteration**: Cycling through a **code block** multiple times. 

**logical operator**: An **operator** specific to a **boolean** **expression**. They are: ```and```, ```or```, and ```not```.

**method**: A function that is attached to a specific object. For example, the string method ```upper()``` capitilizes the string. The code ```print('a'.upper())``` will print an uppercase "A."

**nesting**: When a **code block** is included inside another **code block**. For example, when an ```if``` statement  has another ```if``` statement within it. 

**operator**: Special symbols used in expresions, such as the multiply symbol ```*```, the addition symbol ```+``` or the greater than symbol ```>```, the word ```and```. 

**parameter**: A variable name declared in the function definition to access an argument passed to the function. A function cna have zero, one, or multiple parameters.

**program**: Written computer code, such as in Python, that performs a series of instructions.

**return value** The value returned by a function with the ```return``` **statement**.

**rules of precedence**: The rules that govern the order that **operators** are **evaluated**. For instance, multiplcation such as ```3 * 2``` is **evaluated** before addition such as ```2 + 1```. Thus, the expression ```2 + 3 * 2``` evaluates to ```8```. A good resource is this [Operator precedence table](http://interactivepython.org/runestone/static/CS152f17/Appendices/PrecedenceTable.html) from "How To Think Like a Computer Scientist"


**runtime error**: The program works initially, but certain input or changes in values cause it to crash and throw an error. Example, accidently dividing by zero. 

**semantic error**: The program continues to run (does not crash), but the results are unexpected.

**state**: (Needs a better explanation.) All the values of every variable in the system at a moment in time. Debugging often involves stepping through a program line-by-line and updating how each line changes the state of the system.

**statement**: An instruction run by Python. For instance, the ```if``` statement or a ```while``` loop. (```while``` is a statement that is run in a loop until it **evaluates** to ```False```)

**variable**: A name defined by a programmer that referes to a value of a specific **data type**. For example: ```my_int = 5``` assigns the variable ```my_int``` the integer ```5```.
